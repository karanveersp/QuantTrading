```python
import pandas as pd

df = pd.read_csv("BTCUSD-OneHour-11-1-23-till-2-27-24.csv", parse_dates=["Key"], index_col="Key")
df.index = pd.to_datetime(df.index, format="mixed")
df.index = df.index.tz_localize('UTC').tz_convert('America/New_York')
```

```python
    
import plotly.express as px
import plotly.graph_objects as go


def plot_close_prices_trades(day, df):
    df["Day"] = df.index.day_name()
    # Assuming 'df' is your DataFrame with a DateTimeIndex and a 'Price' column
    # First, ensure the index is localized to UTC and then convert to EST
    monday_df = df[df["Day"] == day].copy()
    monday_df["NormalizedTime"] = monday_df.index.map(lambda dt: pd.Timestamp(year=2000, month=1, day=1, hour=dt.hour, minute=dt.minute, second=dt.second))
    monday_df["PriceChange"] = monday_df["Close"].pct_change()
    monday_df.dropna(inplace=True)
    monday_df = monday_df[monday_df["NormalizedTime"] > pd.Timestamp(2000, 1, 1, 00, 00, 00, tz="America/New_York")]
    fig = px.bar(monday_df, x=monday_df["NormalizedTime"], y=monday_df["PriceChange"], title=f'{day} PriceChange')
    fig.show()

plot_close_prices_trades("Monday", df)
plot_close_prices_trades("Tuesday", df)
plot_close_prices_trades("Wednesday", df)
plot_close_prices_trades("Thursday", df)
plot_close_prices_trades("Friday", df)
plot_close_prices_trades("Saturday", df)

```

```python

# Assuming 'df' includes a 'Volume' column
avg_volume_by_day = df.groupby('Day')['Volume'].mean().reset_index()
fig = px.bar(avg_volume_by_day, x='Day', y='Volume', title='Average Trading Volume by Day of the Week')
fig.show()
```
